```python
#even numbers from 10 to 50
b=1
while b<=51:
    if b%2==0:
        print(b)
    b=b+1
#printing numbers from 1 to 10
x=1
while x<=10:
  print(x)
  x=x+1

# 5 table using while loop'
n=5
c=1
while c<=10:
    print(n,"x",c,"=",n*c)
    c+=1

#sum of digits
num = 54321
digit_sum = 0
while num > 0:
    digit = num % 10
    digit_sum += digit
    num //= 10
print("Sum of digits:", digit_sum)

# 'reversing num using while'
s=153
res=0
while s>0:
    digit=s%10

    res=res*10+digit
    s//=10
print(res)

#'fibonacci series'
a,b=0,1
n=0
while n<=11:
    print(a)
    a,b=b,a+b
    n+=1

#'num is palindrome or not'
num = 121
original = num
res= 0
while num > 0:
    digit = num % 10
    res = res* 10 + digit
    num = num // 10

if original == res:
    print("Palindrome")
else:
    print("Not a palindrome")
```
